```python
from datasets import load_dataset
qna_dataset = load_dataset("sadeem-ai/arabic-qna")
```

```python
qna_dataset
```

```python
qna_dataset=qna_dataset.filter(lambda x:x['has_answer']==True)
```

```python
qna_dataset
```

```python
doc_text=qna_dataset['train']['text']
```

```python
meta_data=[
    {
        "source":rec['source'],
        "title":rec['title'],
    }
    for rec in qna_dataset['train']
]
```

```python
meta_data[50]
```

```python
doc_ids=[
    str(i)
    for i in range(len(doc_text))
]
```

```python
from sentence_transformers import SentenceTransformer
model_id="sentence-transformers/distiluse-base-multilingual-cased-v1"
device="cuda:0"
model = SentenceTransformer(model_id, device=device)
text="السلام عليكم و رحمة الله وبركاته"
embeddings = model.encode(text)
print(embeddings)
```

```python
encoded_docs=model.encode(doc_text,show_progress_bar=True)
```

```python
encoded_docs.shape
```

```python
import chromadb
```

```python
chroma_client = chromadb.PersistentClient(path="./chroma_db-ar-docs")
```

```python
collection=chroma_client.create_collection(
    name="ar-docs",
    metadata={"hnsw:space": "cosine"}
    )
```

```python
collection.add(
    ids=doc_ids,
    documents=doc_text,
    embeddings=encoded_docs,
    metadatas=meta_data
)
```

```python
question="ما هي ممارسة"
embed_q=model.encode(question)
reult=collection.query(
    query_embeddings=embed_q.tolist(),
    n_results=3
)
```

```python
reult
```

# faiss


```python
import faiss
import numpy as np
from copy import deepcopy
```

```python
norm_encoded_docs=deepcopy(encoded_docs)
faiss.normalize_L2(norm_encoded_docs)
faiss_index=faiss.IndexIDMap(faiss.IndexFlatIP(512))
faiss_index.add_with_ids(norm_encoded_docs,doc_ids)
```

```python
question="ما يمكن أن يؤدي المكان الذي ينتمي إليه الشخص إليه"

question_embed=model.encode([question])
faiss.normalize_L2(question_embed)
result=faiss_index.search(question_embed,3)
```
